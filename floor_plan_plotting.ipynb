```python
import fitz  # PyMuPDF

def pdf_to_png(input_pdf, output_folder):
    pdf_document = fitz.open(input_pdf)

    for page_number in range(pdf_document.page_count):
        page = pdf_document.load_page(page_number)
        image_matrix = fitz.Matrix(3, 3)  # Increase the DPI for higher resolution images
        image = page.get_pixmap(matrix=image_matrix)
        image.save(f"{output_folder}/page_{page_number+1}.png")

    pdf_document.close()

        
# Example usage:
input_pdf_path = 'C:/Users/ali/Downloads/Compilation_plan_amenagement_7Rue_Chateaudun.pdf'
output_folder = 'C:/Users/ali/SmartBuilding/'
pdf_to_png(input_pdf_path, output_folder)
```
