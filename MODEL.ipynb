•	Utiliser des modèles de séries chronologiques pour détecter les anomalies dans les schémas d'occupation, ce qui pourrait indiquer un comportement ou des événements inhabituels dans le bâtiment.
•	Meilleur choix : la fonction de détection d'anomalies de Prophet. Il dispose d'un support intégré pour la détection d'anomalies, ce qui peut être utile pour identifier des modèles d'occupation inhabituels.


```python
import pandas as pd
from prophet import Prophet

data = merged_df.rename(columns={'timestamp [dd/mm/yyyy HH:MM]': 'ds', 'ki [0:vacant 1:occupied]': 'y'})

model = Prophet(interval_width=0.95, daily_seasonality=False, yearly_seasonality=False)
model.add_country_holidays(country_name='AT')
model.add_seasonality(name='custom', period=24 * 60, fourier_order=3)
model.add_regressor('gh [W/m2]')
model.add_regressor('tempOut [C]')
model.add_regressor('rh [%]')
model.add_regressor('wind speed [m/s]')
model.add_regressor('Wind direction [Degree] [North:0 East:90 South:180 West:270]')
model.fit(data)

forecast = model.predict(data)
forecast['anomaly'] = 0
forecast.loc[forecast['yhat'] < forecast['yhat_lower'], 'anomaly'] = 1
forecast.loc[forecast['yhat'] > forecast['yhat_upper'], 'anomaly'] = 1

import matplotlib.pyplot as plt

plt.plot(data['ds'], data['y'], label='Occupation réelle', color='blue')
plt.scatter(forecast.loc[forecast['anomaly'] == 1, 'ds'], forecast.loc[forecast['anomaly'] == 1, 'y'], color='red', label='Anomalies')
plt.legend()
plt.show()
```

```python
import holidays
def filter_work_hour_drop_holidays(merged_df,features,target):
    df = merged_df[features+[target]+['timestamp [dd/mm/yyyy HH:MM]']]
    print(df.describe())
    df.info()

    # Convert the timestamp column to a proper datetime format (assuming the column name is 'timestamp')
    df['timestamp'] = pd.to_datetime(df['timestamp [dd/mm/yyyy HH:MM]'])

    # Extract day and hour information from the 'timestamp' column
    # working_hours_df['day'] = df['timestamp'].dt.day
    df['minute'] = df['timestamp'].dt.minute
    df['hour'] = df['timestamp'].dt.hour
    df['day'] = df['timestamp'].dt.day
    df['month'] = df['timestamp'].dt.month
    df['year'] = df['timestamp'].dt.year
    # working_hours_df['hour'] = df['timestamp'].dt.hour

    # Filter the DataFrame to include only rows during working hours (8:00 AM to 6:00 PM)
    df = df[(df['timestamp'].dt.hour >= 8) & (df['timestamp'].dt.hour < 18)]

    # Identify weekends and public holidays in the US
    us_holidays = holidays.US(years=int(df['timestamp'].dt.year.min()), observed=True)
    df['is_weekend'] = df['timestamp'].dt.dayofweek.isin([5, 6])  # 5=Saturday, 6=Sunday
    df['is_public_holiday'] = df['timestamp'].dt.date.isin(us_holidays)

    # Filter out weekends and public holidays from the analysis
    df = df[~(df['is_weekend'] | df['is_public_holiday'])]
    
    
    
    # Create a list of Austrian holidays
    at_holidays = holidays.Austria(years=range(2013, 2014))

    # Function to check if a date is a holiday or a weekend (Saturday or Sunday)
    def is_weekend_or_holiday(date):
        return date.weekday() >= 5 or date in at_holidays

    # Apply the mask to filter out weekends and holidays
    df = df[~df['timestamp'].apply(is_weekend_or_holiday)]
    # df[target+'occupancy_absence'] = df [target].apply(categorize_presence)
    
    return df
```

```python
# Times Series : LSTM, MLP, CNN, RNN 

import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from keras.layers import LSTM, Dense, RepeatVector, TimeDistributed
from keras.models import Sequential

# Charger les données (assurez-vous d'avoir vos données préparées sous forme de DataFrame)
# features = ['ki [%]', 'ki [C]', 'gh [W/m2]', 'ki  [1:closed 0:open]', 'tempOut [C]', 'rh [%]','ki [0:off 1:on]', 'wind speed [m/s]']
features = ['o4 [%]', 'o4 [C]', 'gh [W/m2]', 'o4_1 [1:closed 0:open]','o4_2 [1:closed 0:open]','tempOut [C]', 'rh [%]','o4_1 [0:off 1:on]', 'o4_2 [0:off 1:on]', 'wind speed [m/s]']
target = ["o4  [0:vacant 1:occupied]"]

# Assuming your 'timestamp [dd/mm/yyyy HH:MM]' column is in datetime format
merged_df['timestamp'] = pd.to_datetime(merged_df['timestamp [dd/mm/yyyy HH:MM]'])

# Creates time series features from datetime index
# Set the timestamp as the DataFrame index (you missed this)
merged_df.set_index('timestamp', inplace=True)

def create_features(df, label=None):
    # Creates time series features from datetime index
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.dayofweek
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofyear'] = df.index.dayofyear
    df['dayofmonth'] = df.index.day
    df['weekofyear'] = df.index.isocalendar().week
    
    # X = df[['hour','dayofweek','quarter','month','year','dayofyear','dayofmonth','weekofyear','ki [%]', 'ki [C]', 'gh [W/m2]', 
    #                'ki  [1:closed 0:open]', 'tempOut [C]', 'rh [%]','ki [0:off 1:on]', 'wind speed [m/s]']] 
    
    X = df[features]
   

    if label:
        y = df[label]        
        return X, y
    return X


# Filter data based on a specific split date
split_date = '2013-10-10'

# diviser les données en apprentissage[train] et test 
train_data = merged_df.loc[merged_df.index <= split_date].copy()
test_data = merged_df.loc[merged_df.index > split_date].copy()
print(test_data.shape)

X_train, y_train = create_features(train_data, label="o4 [0:vacant 1:occupied]")
X_test, y_test = create_features(test_data, label="o4 [0:vacant 1:occupied]")    


# Normalisation des données
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Fonction pour créer des séquences temporelles
def create_sequences(data, target, sequence_length):
    X_sequences = []
    y_sequences = []

    for i in range(len(data) - sequence_length):
        X_seq = data[i:i+sequence_length]
        y_seq = target[i+sequence_length-1:i+sequence_length]  # Utilisez le dernier élément comme cible

        X_sequences.append(X_seq)
        y_sequences.append(y_seq)

    return np.array(X_sequences), np.array(y_sequences)


sequence_length = 10 # Longueur de la séquence temporelle
X_train_seq, y_train_seq = create_sequences(X_train, y_train, sequence_length)
X_test_seq, y_test_seq = create_sequences(X_test, y_test, sequence_length)

y_train_seq = y_train_seq.reshape(-1, 1)
y_test_seq = y_test_seq.reshape(-1, 1)


# Création du modèle LSTM
# model = keras.Sequential([
#     keras.layers.LSTM(units=128, activation='relu', input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
#     keras.layers.Dense(32, 'relu', kernel_initializer='he_uniform'),
#     # keras.layers.Dense(16, 'relu'),
#     keras.layers.Dense(1, 'sigmoid')
# ])

model = keras.Sequential(
    # keras.layers.LSTM(units=128, activation='relu', input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
    # keras.layers.Dense(32, 'relu', kernel_initializer='he_uniform'),
    keras.layers.Input(shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
                keras.layers.Bidirectional(keras.layers.LSTM(1024, return_sequences=True)),
                keras.layers.Bidirectional(keras.layers.LSTM(512, return_sequences=True)),
                keras.layers.Bidirectional(keras.layers.LSTM(256, return_sequences=True)),
                keras.layers.Bidirectional(keras.layers.LSTM(128, return_sequences=True)),
                keras.layers.Dense(128, activation='selu'),
                keras.layers.Dense(1)
    # keras.layers.Dense(16, 'relu'),
    # keras.layers.Dense(1, 'sigmoid')
)

# model = keras.Sequential([
#     keras.layers.LSTM(units=64, activation='relu', input_shape=(X_train_seq.shape[1], X_train_seq.shape[2]), return_sequences=False),
#     keras.layers.LSTM(32, 'relu', kernel_initializer='he_uniform',return_sequences=False),
#     keras.layers.Dropout(0.2),
#     # keras.layers.Dense(16, 'relu'),
#     keras.layers.Dense(1, 'sigmoid')
# ])



# model.summary()

model.compile(optimizer ="adam",loss="mean_squared_error",metrics=['accuracy'])

# Training the model
history = model.fit(X_train_seq, y_train_seq, epochs=10, batch_size=32, validation_data=(X_test_seq, y_test_seq))

# Prediction
y_pred_seq = model.predict(X_test_seq)

# Reshape the predictions to match the original shape
# y_pred_seq = y_pred_seq.reshape(-1, 1)

# Visualize the results
plt.plot(y_test_seq, label='True Value')
plt.plot(y_pred_seq, label='Prediction')
plt.legend()
plt.show()


```

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Bidirectional


input_dim = X_train.shape[1]  # Replace this with the actual number of input features
X_val, y_val = X_test_seq, y_test_seq

# Créer une liste de modèles
models = [
    # Sequential([
    #     Dense(64, input_dim=input_dim, activation='relu'),
    #     Dense(32, activation='relu'),
    #     Dense(1, activation='sigmoid')
    # ]),
    Sequential([
        Conv1D(32, kernel_size=3, activation='relu', input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
        MaxPooling1D(pool_size=2),
        Flatten(),
        Dense(1, activation='sigmoid')
    ]),
    Sequential([
        LSTM(64, input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
        Dense(1, activation='sigmoid')
    ]),
    Sequential([
        Bidirectional(LSTM(64), input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
        Dense(1, activation='sigmoid')
    ])
]

# Créer des listes pour stocker les métriques
mse_scores = []
rmse_scores = []
r_squared_scores = []

# Boucle à travers les modèles
for model in models:
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mse'])
    history = model.fit(X_train_seq, y_train_seq, epochs=10, batch_size=32, validation_data=(X_test_seq, y_test_seq))
    
    # Évaluer le modèle sur l'ensemble de test
    loss, mse = model.evaluate(X_test_seq, y_test_seq)
    
    # Calculer RMSE et R-squared
    y_pred = model.predict(X_test_seq)
    rmse = np.sqrt(mean_squared_error(X_test_seq, y_pred))
    r_squared = 1 - (mean_squared_error(X_test_seq, y_pred) / np.var(X_test_seq))
    
    # Enregistrer les métriques dans les listes
    mse_scores.append(mse)
    rmse_scores.append(rmse)
    r_squared_scores.append(r_squared)

# Trouver le meilleur modèle en fonction de la métrique souhaitée (par exemple, MSE)
best_model_index = np.argmin(mse_scores)
best_model = models[best_model_index]

# Afficher les métriques pour chaque modèle
for i, model in enumerate(models):
    print(f"Model {i+1}:")
    print(f"MSE: {mse_scores[i]}")
    print(f"RMSE: {rmse_scores[i]}")
    print(f"R-squared: {r_squared_scores[i]}")
    print()

# Afficher les métriques du meilleur modèle
print("Best Model (based on MSE):")
print(f"MSE: {mse_scores[best_model_index]}")
print(f"RMSE: {rmse_scores[best_model_index]}")
print(f"R-squared: {r_squared_scores[best_model_index]}")

# Entraîner le meilleur modèle sur l'ensemble complet de données
best_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mse'])
best_history = best_model.fit(X_train_seq, y_train_seq, epochs=10, batch_size=32, validation_data=(X_test_seq, y_test_seq))

# Afficher les courbes d'apprentissage pour le meilleur modèle
plt.plot(best_history.history['mse'], label='Training MSE')
plt.plot(best_history.history['val_mse'], label='Validation MSE')
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.legend()
plt.show()
```
