•	Utiliser des modèles de séries chronologiques pour détecter les anomalies dans les schémas d'occupation, ce qui pourrait indiquer un comportement ou des événements inhabituels dans le bâtiment.
•	Meilleur choix : la fonction de détection d'anomalies de Prophet. Il dispose d'un support intégré pour la détection d'anomalies, ce qui peut être utile pour identifier des modèles d'occupation inhabituels.


```python
import pandas as pd
from prophet import Prophet

data = merged_df.rename(columns={'timestamp [dd/mm/yyyy HH:MM]': 'ds', 'ki [0:vacant 1:occupied]': 'y'})

model = Prophet(interval_width=0.95, daily_seasonality=False, yearly_seasonality=False)
model.add_country_holidays(country_name='AT')
model.add_seasonality(name='custom', period=24 * 60, fourier_order=3)
model.add_regressor('gh [W/m2]')
model.add_regressor('tempOut [C]')
model.add_regressor('rh [%]')
model.add_regressor('wind speed [m/s]')
model.add_regressor('Wind direction [Degree] [North:0 East:90 South:180 West:270]')
model.fit(data)

forecast = model.predict(data)
forecast['anomaly'] = 0
forecast.loc[forecast['yhat'] < forecast['yhat_lower'], 'anomaly'] = 1
forecast.loc[forecast['yhat'] > forecast['yhat_upper'], 'anomaly'] = 1

import matplotlib.pyplot as plt

plt.plot(data['ds'], data['y'], label='Occupation réelle', color='blue')
plt.scatter(forecast.loc[forecast['anomaly'] == 1, 'ds'], forecast.loc[forecast['anomaly'] == 1, 'y'], color='red', label='Anomalies')
plt.legend()
plt.show()
```

## Evaluation Apprentissage : 

On contstate que les methodes d'ensembles d'apprentissage supérvisée est les plus performantes : les arbres de classification tel que les forêts d'arbres aléatoires, Le boosting d'arbre de classification; améliorent la qualité ou la fiabilité de la prédiction en construisant plusieurs arbres de décision depuis les données.

Définition des arbres de décision : 

C'est une technique d'apprentissage supervisé : on utilise un ensemble de données pour lesquelles on connaît la valeur de la variable-cible afin de construire l'arbre (données dites étiquetées), puis on extrapole les résultats à l'ensemble des données de test. Les arbres de décision font partie des algorithmes les plus populaires en apprentissage automatique.[Arbre de décision wiki](https://fr.wikipedia.org/wiki/Arbre_de_d%C3%A9cision_(apprentissage)#cite_note-11)

Pour le modéle 2 : la prédiction d'occupation de l'open Desk 4 : l'exactitude de prédiction est 93%, 93%, 91% et 89% respectivement pour le Random forest, Extra Trees Classifier, l'arbre de décision et le Boosting.
Par contre le random forest et Extra Trees Classifier prend beaucoup plus de temps : 4.17 Sec et 2.01 Sec.

Pycaret utilise l’inférence conditionnelle pour construire des arbres de décision plus robustes en utilisant des tests de signification pour sélectionner les variables à chaque étape de la construction de l’arbre. Cette approche permet d’éviter le biais de sélection de variables qui peut affecter les méthodes traditionnelles d’arbre de décision et le probléme de sur-apprentissage. Cependant, l’inférence conditionnelle peut être plus coûteuse en temps de calcul que les méthodes traditionnelles.

#### Time Serie forcasting :

Prophet : Prophet est une procédure de prévision des données de séries chronologiques basée sur un modèle additif dans lequel les tendances non linéaires sont adaptées à la saisonnalité annuelle, hebdomadaire et quotidienne, ainsi qu'aux effets des vacances. Cela fonctionne mieux avec des séries chronologiques ayant de forts effets saisonniers et plusieurs saisons de données historiques. Prophet est robuste aux données manquantes et aux changements de tendance, et gère généralement bien les valeurs aberrantes.

(Prophet web site)[https://facebook.github.io/prophet/] 

Importation des bibliothèques nécessaires : On importe les bibliothèques pandas et Prophet, qui sont utilisées pour la manipulation de données et la prédiction de séries temporelles respectivement.

Préparation des données : Les colonnes de la DataFrame humidity sont renommées pour correspondre aux noms de colonnes que la bibliothèque Prophet attend ('ds' pour la date et 'y' pour la variable à prédire). La colonne des dates est convertie au format datetime.

Division des données en ensemble d'entraînement et de test : Les données sont divisées en un ensemble d'entraînement (toutes les données jusqu'à la date '2013-10-01') et un ensemble de test (toutes les données après cette date).

Définition de la période initiale et de l'horizon : La période initiale (la durée de l'ensemble d'entraînement) et l'horizon (la durée de l'ensemble de test) sont définis.

Création et ajustement du modèle : Un modèle Prophet est créé avec une certaine configuration. Le modèle est ensuite ajusté sur l'ensemble d'entraînement, les paramètres du modèle Prophet utilisés  : 
    
    - changepoint_prior_scale: Ce paramètre contrôle la flexibilité du modèle pour s'adapter aux changements. Des valeurs plus élevées rendent le modèle plus flexible pour s'adapter à ces changements. Dans votre cas, la valeur est définie à '0.010'.

    - seasonality_prior_scale: C'est un paramètre qui contrôle la force de la saisonnalité. Si vous pensez que les effets de saisonnalité sont forts, vous pouvez augmenter ce paramètre. Ici, il est défini à '0.01'.

    - seasonality_mode: Il existe deux modes de saisonnalité : 'additive' et 'multiplicative'. Le mode 'multiplicative' est approprié lorsque la saisonnalité augmente avec la tendance, tandis que le mode 'additive' est approprié lorsque la saisonnalité est plus ou moins la même quelle que soit la tendance. Dans votre cas, le mode 'multiplicative' est utilisé.

    - yearly_seasonality et weekly_seasonality: Ces paramètres permettent au modèle de prendre en compte la saisonnalité annuelle et hebdomadaire. Dans votre cas, ils sont tous deux définis sur True, ce qui signifie que le modèle prendra en compte les effets saisonniers annuels et hebdomadaires.

    - add_country_holidays: Cette méthode permet d'ajouter les jours fériés d'un pays spécifique dans le modèle. Ici, vous avez ajouté les jours fériés de l'Autriche (code du pays 'AT').

Prédiction : Un DataFrame 'future', qui contient toutes les dates de la période actuelle jusqu'à 90 jours est créé pour les dates futures pour lesquelles nous voulons faire des prédictions. Ensuite, le modèle fait des prédictions pour pour chaque heure des 90 prochains jours. Il retourne un DataFrame qui contient les prédictions ('yhat') ainsi que les intervalles de confiance inférieurs et supérieurs ('yhat_lower' et 'yhat_upper') pour ces prédictions.

Visualisation des prédictions : Les prédictions sont visualisées à l'aide de la méthode plot de Prophet : génère une visualisation de la série temporelle qui inclut les valeurs historiques et les prédictions futures. La visualisation affiche les composantes suivantes :

        - Les points noirs représentent les données observées.
        - La ligne bleue représente la prédiction (ou la tendance) du modèle Prophet.
        - La zone bleu clair représente l'intervalle de confiance, qui est l'incertitude de la prédiction.
Aussi, une visualisation qui décompose les prédictions en plusieurs composantes :

Tendance : Cette partie montre la tendance sous-jacente de la série temporelle, c'est-à-dire la tendance à long terme sans les effets saisonniers ou les jours fériés.

Saisonnalité annuelle, hebdomadaire et quotidienne : Ces parties montrent comment la série temporelle est affectée par la saisonnalité annuelle et hebdomadaire. Par exemple, dans le graphique de saisonnalité hebdomadaire, vous pouvez voir comment les prédictions varient en fonction du jour de la semaine.

Effets des jours fériés : Si vous avez ajouté des jours fériés à votre modèle (comme c'est le cas ici avec m.add_country_holidays(country_name='AT')), cette partie montrera l'impact des jours fériés sur les prédictions.

Évaluation du modèle : Le modèle est évalué en utilisant la validation croisée. La performance du modèle est évaluée en fonction de différentes mesures telles que le MAE (Mean Absolute Error), le RMSE (Root Mean Squared Error), etc. Ces mesures sont ensuite affichées. 
     les définitions des métriques mentionnées :

        MAE (Mean Absolute Error) : C'est la moyenne des différences absolues entre les prédictions et les valeurs réelles. Cela donne une idée de l'ampleur des erreurs sans tenir compte de leur direction. Une valeur de MAE plus faible indique de meilleures performances du modèle.

        RMSE (Root Mean Squared Error) : C'est la racine carrée de la moyenne des carrés des erreurs entre les prédictions et les valeurs réelles. Contrairement à la MAE, le RMSE donne plus de poids aux erreurs plus grandes. C'est-à-dire que le RMSE augmente de manière disproportionnée pour les grandes erreurs. Un RMSE plus faible indique de meilleures performances du modèle.

        En plus de ces deux, il existe plusieurs autres métriques de performance couramment utilisées pour évaluer les modèles de prédiction :

        MSE (Mean Squared Error) : C'est la moyenne des carrés des erreurs entre les prédictions et les valeurs réelles. Il est similaire au RMSE, mais sans prendre la racine carrée. Un MSE plus faible indique de meilleures performances du modèle.

        MAPE (Mean Absolute Percentage Error) : C'est la moyenne des différences absolues entre les prédictions et les valeurs réelles, exprimée en pourcentage. Il donne une idée de l'erreur en termes de pourcentage, ce qui peut être plus intuitif pour certaines applications. Un MAPE plus faible indique de meilleures performances du modèle.

```python
### Prophet : forcasting time series package: sesonality, Holidays, special Events

import pandas as pd
from prophet import Prophet
from prophet.diagnostics import cross_validation
from datetime import datetime
path = "https://raw.githubusercontent.com/alimanager/smartBuilding/master/"
humd=pd.read_csv(f'{path}06_rhu_in.csv')
humd = humd["o4 [%]"]
humd = humd.rename(columns = {'timestamp [dd/mm/yyyy HH:MM]' : 'ds','o4 [%]': 'y'})
humd["ds"]= pd.to_datetime(humd["ds"], format = "%d/%m/%Y %H:%M")

humd.set_index('ds', inplace=True)
humd_hourly = humd.resample('H').mean()  # or use .sum(), .max(), etc. depending on your needs
humd_hourly.reset_index(inplace=True)

#Train & test split data 
Date_split = '2013-10-01'
train = humd[humd['ds'] <= Date_split] 
test = humd[humd['ds'] > Date_split] 

# Define initial and horizon
# Specify the length of the initial training period and the prediction horizon
initial = str((train['ds'].max() - train['ds'].min()).days) + ' days'
horizon = str((test['ds'].max() - test['ds'].min()).days) + ' days'

m = Prophet(changepoint_prior_scale='0.01',seasonality_prior_scale = '0.01',
            seasonality_mode='multiplicative',
            yearly_seasonality=True, weekly_seasonality=True,
            changepoint_range=0.9)
# m = Prophet()
m.add_country_holidays(country_name='AT') ### country_name='FR'for france
m.fit(humd)
# m.fit(humd)

future = m.make_future_dataframe(periods=30*24, freq='H')

# Prédiction & Vizualisation : 
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
fig = m.plot(forecast)
fig = m.plot_components(forecast)

# Evaluation : 
from prophet.diagnostics import performance_metrics
cv = cross_validation(model=m,initial=initial, horizon= horizon,period='90 days')
metrics = performance_metrics(cv)
print(metrics)
```

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Bidirectional


input_dim = X_train.shape[1]  # Replace this with the actual number of input features
X_val, y_val = X_test_seq, y_test_seq

# Créer une liste de modèles
models = [
    # Sequential([
    #     Dense(64, input_dim=input_dim, activation='relu'),
    #     Dense(32, activation='relu'),
    #     Dense(1, activation='sigmoid')
    # ]),
    Sequential([
        Conv1D(32, kernel_size=3, activation='relu', input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
        MaxPooling1D(pool_size=2),
        Flatten(),
        Dense(1, activation='sigmoid')
    ]),
    Sequential([
        LSTM(64, input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
        Dense(1, activation='sigmoid')
    ]),
    Sequential([
        Bidirectional(LSTM(64), input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
        Dense(1, activation='sigmoid')
    ])
]

# Créer des listes pour stocker les métriques
mse_scores = []
rmse_scores = []
r_squared_scores = []

# Boucle à travers les modèles
for model in models:
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mse'])
    history = model.fit(X_train_seq, y_train_seq, epochs=10, batch_size=32, validation_data=(X_test_seq, y_test_seq))
    
    # Évaluer le modèle sur l'ensemble de test
    loss, mse = model.evaluate(X_test_seq, y_test_seq)
    
    # Calculer RMSE et R-squared
    y_pred = model.predict(X_test_seq)
    rmse = np.sqrt(mean_squared_error(X_test_seq, y_pred))
    r_squared = 1 - (mean_squared_error(X_test_seq, y_pred) / np.var(X_test_seq))
    
    # Enregistrer les métriques dans les listes
    mse_scores.append(mse)
    rmse_scores.append(rmse)
    r_squared_scores.append(r_squared)

# Trouver le meilleur modèle en fonction de la métrique souhaitée (par exemple, MSE)
best_model_index = np.argmin(mse_scores)
best_model = models[best_model_index]

# Afficher les métriques pour chaque modèle
for i, model in enumerate(models):
    print(f"Model {i+1}:")
    print(f"MSE: {mse_scores[i]}")
    print(f"RMSE: {rmse_scores[i]}")
    print(f"R-squared: {r_squared_scores[i]}")
    print()

# Afficher les métriques du meilleur modèle
print("Best Model (based on MSE):")
print(f"MSE: {mse_scores[best_model_index]}")
print(f"RMSE: {rmse_scores[best_model_index]}")
print(f"R-squared: {r_squared_scores[best_model_index]}")

# Entraîner le meilleur modèle sur l'ensemble complet de données
best_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mse'])
best_history = best_model.fit(X_train_seq, y_train_seq, epochs=10, batch_size=32, validation_data=(X_test_seq, y_test_seq))

# Afficher les courbes d'apprentissage pour le meilleur modèle
plt.plot(best_history.history['mse'], label='Training MSE')
plt.plot(best_history.history['val_mse'], label='Validation MSE')
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.legend()
plt.show()
```
