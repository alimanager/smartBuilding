```python
# Filter Weekends and Holidays : 
import holidays
def filter_work_hour_drop_holidays(merged_df,features,target):
    df = merged_df[features+[target]+['timestamp [dd/mm/yyyy HH:MM]']]
    print(df.describe())
    df.info()

    # Convert the timestamp column to a proper datetime format (assuming the column name is 'timestamp')
    df['timestamp'] = pd.to_datetime(df['timestamp [dd/mm/yyyy HH:MM]'])

    # Extract day and hour information from the 'timestamp' column
    # working_hours_df['day'] = df['timestamp'].dt.day
    df['minute'] = df['timestamp'].dt.minute
    df['hour'] = df['timestamp'].dt.hour
    df['day'] = df['timestamp'].dt.day
    df['month'] = df['timestamp'].dt.month
    df['year'] = df['timestamp'].dt.year
    # working_hours_df['hour'] = df['timestamp'].dt.hour

    # Filter the DataFrame to include only rows during working hours (8:00 AM to 6:00 PM)
    df = df[(df['timestamp'].dt.hour >= 8) & (df['timestamp'].dt.hour < 18)]

    # Identify weekends and public holidays in the US
    us_holidays = holidays.US(years=int(df['timestamp'].dt.year.min()), observed=True)
    df['is_weekend'] = df['timestamp'].dt.dayofweek.isin([5, 6])  # 5=Saturday, 6=Sunday
    df['is_public_holiday'] = df['timestamp'].dt.date.isin(us_holidays)

    # Filter out weekends and public holidays from the analysis
    df = df[~(df['is_weekend'] | df['is_public_holiday'])]

    # Create a list of Austrian holidays
    at_holidays = holidays.Austria(years=range(2013, 2014))

    # Function to check if a date is a holiday or a weekend (Saturday or Sunday)
    def is_weekend_or_holiday(date):
        return date.weekday() >= 5 or date in at_holidays

    # Apply the mask to filter out weekends and holidays
    df = df[~df['timestamp'].apply(is_weekend_or_holiday)]
    # df[target+'occupancy_absence'] = df [target].apply(categorize_presence)
    
    return df
```

```python
import pandas as pd
from prophet import Prophet

data = merged_df.rename(columns={'timestamp [dd/mm/yyyy HH:MM]': 'ds', 'ki [0:vacant 1:occupied]': 'y'})

model = Prophet(interval_width=0.95, daily_seasonality=False, yearly_seasonality=False)
model.add_country_holidays(country_name='AT')
model.add_seasonality(name='custom', period=24 * 60, fourier_order=3)
model.add_regressor('gh [W/m2]')
model.add_regressor('tempOut [C]')
model.add_regressor('rh [%]')
model.add_regressor('wind speed [m/s]')
model.add_regressor('Wind direction [Degree] [North:0 East:90 South:180 West:270]')
model.fit(data)

forecast = model.predict(data)
forecast['anomaly'] = 0
forecast.loc[forecast['yhat'] < forecast['yhat_lower'], 'anomaly'] = 1
forecast.loc[forecast['yhat'] > forecast['yhat_upper'], 'anomaly'] = 1

import matplotlib.pyplot as plt

plt.plot(data['ds'], data['y'], label='Occupation relle', color='blue')
plt.scatter(forecast.loc[forecast['anomaly'] == 1, 'ds'], forecast.loc[forecast['anomaly'] == 1, 'y'], color='red', label='Anomalies')
plt.legend()
plt.show()
```

#### Time Serie forcasting :

Prophet : Prophet est une procdure de prvision des donnes de sries chronologiques base sur un modle additif dans lequel les tendances non linaires sont adaptes  la saisonnalit annuelle, hebdomadaire et quotidienne, ainsi qu'aux effets des vacances. Cela fonctionne mieux avec des sries chronologiques ayant de forts effets saisonniers et plusieurs saisons de donnes historiques. Prophet est robuste aux donnes manquantes et aux changements de tendance, et gre gnralement bien les valeurs aberrantes.

(Prophet web site)[https://facebook.github.io/prophet/] 

Importation des bibliothques ncessaires : On importe les bibliothques pandas et Prophet, qui sont utilises pour la manipulation de donnes et la prdiction de sries temporelles respectivement.

Prparation des donnes : Les colonnes de la DataFrame humidity sont renommes pour correspondre aux noms de colonnes que la bibliothque Prophet attend ('ds' pour la date et 'y' pour la variable  prdire). La colonne des dates est convertie au format datetime.

Division des donnes en ensemble d'entranement et de test : Les donnes sont divises en un ensemble d'entranement (toutes les donnes jusqu' la date '2013-10-01') et un ensemble de test (toutes les donnes aprs cette date).

Dfinition de la priode initiale et de l'horizon : La priode initiale (la dure de l'ensemble d'entranement) et l'horizon (la dure de l'ensemble de test) sont dfinis.

Cration et ajustement du modle : Un modle Prophet est cr avec une certaine configuration. Le modle est ensuite ajust sur l'ensemble d'entranement, les paramtres du modle Prophet utiliss  : 
    
    - changepoint_prior_scale: Ce paramtre contrle la flexibilit du modle pour s'adapter aux changements. Des valeurs plus leves rendent le modle plus flexible pour s'adapter  ces changements. Dans votre cas, la valeur est dfinie  '0.010'.

    - seasonality_prior_scale: C'est un paramtre qui contrle la force de la saisonnalit. Si vous pensez que les effets de saisonnalit sont forts, vous pouvez augmenter ce paramtre. Ici, il est dfini  '0.01'.

    - seasonality_mode: Il existe deux modes de saisonnalit : 'additive' et 'multiplicative'. Le mode 'multiplicative' est appropri lorsque la saisonnalit augmente avec la tendance, tandis que le mode 'additive' est appropri lorsque la saisonnalit est plus ou moins la mme quelle que soit la tendance. Dans votre cas, le mode 'multiplicative' est utilis.

    - yearly_seasonality et weekly_seasonality: Ces paramtres permettent au modle de prendre en compte la saisonnalit annuelle et hebdomadaire. Dans votre cas, ils sont tous deux dfinis sur True, ce qui signifie que le modle prendra en compte les effets saisonniers annuels et hebdomadaires.

    - add_country_holidays: Cette mthode permet d'ajouter les jours fris d'un pays spcifique dans le modle. Ici, vous avez ajout les jours fris de l'Autriche (code du pays 'AT').

Prdiction : Un DataFrame 'future', qui contient toutes les dates de la priode actuelle jusqu' 90 jours est cr pour les dates futures pour lesquelles nous voulons faire des prdictions. Ensuite, le modle fait des prdictions pour pour chaque heure des 90 prochains jours. Il retourne un DataFrame qui contient les prdictions ('yhat') ainsi que les intervalles de confiance infrieurs et suprieurs ('yhat_lower' et 'yhat_upper') pour ces prdictions.

Visualisation des prdictions : Les prdictions sont visualises  l'aide de la mthode plot de Prophet : gnre une visualisation de la srie temporelle qui inclut les valeurs historiques et les prdictions futures. La visualisation affiche les composantes suivantes :

        - Les points noirs reprsentent les donnes observes.
        - La ligne bleue reprsente la prdiction (ou la tendance) du modle Prophet.
        - La zone bleu clair reprsente l'intervalle de confiance, qui est l'incertitude de la prdiction.
Aussi, une visualisation qui dcompose les prdictions en plusieurs composantes :

Tendance : Cette partie montre la tendance sous-jacente de la srie temporelle, c'est--dire la tendance  long terme sans les effets saisonniers ou les jours fris.

Saisonnalit annuelle, hebdomadaire et quotidienne : Ces parties montrent comment la srie temporelle est affecte par la saisonnalit annuelle et hebdomadaire. Par exemple, dans le graphique de saisonnalit hebdomadaire, vous pouvez voir comment les prdictions varient en fonction du jour de la semaine.

Effets des jours fris : Si vous avez ajout des jours fris  votre modle (comme c'est le cas ici avec m.add_country_holidays(country_name='AT')), cette partie montrera l'impact des jours fris sur les prdictions.

valuation du modle : Le modle est valu en utilisant la validation croise. La performance du modle est value en fonction de diffrentes mesures telles que le MAE (Mean Absolute Error), le RMSE (Root Mean Squared Error), etc. Ces mesures sont ensuite affiches. 
     les dfinitions des mtriques mentionnes :

        MAE (Mean Absolute Error) : C'est la moyenne des diffrences absolues entre les prdictions et les valeurs relles. Cela donne une ide de l'ampleur des erreurs sans tenir compte de leur direction. Une valeur de MAE plus faible indique de meilleures performances du modle.

        RMSE (Root Mean Squared Error) : C'est la racine carre de la moyenne des carrs des erreurs entre les prdictions et les valeurs relles. Contrairement  la MAE, le RMSE donne plus de poids aux erreurs plus grandes. C'est--dire que le RMSE augmente de manire disproportionne pour les grandes erreurs. Un RMSE plus faible indique de meilleures performances du modle.

        En plus de ces deux, il existe plusieurs autres mtriques de performance couramment utilises pour valuer les modles de prdiction :

        MSE (Mean Squared Error) : C'est la moyenne des carrs des erreurs entre les prdictions et les valeurs relles. Il est similaire au RMSE, mais sans prendre la racine carre. Un MSE plus faible indique de meilleures performances du modle.

        MAPE (Mean Absolute Percentage Error) : C'est la moyenne des diffrences absolues entre les prdictions et les valeurs relles, exprime en pourcentage. Il donne une ide de l'erreur en termes de pourcentage, ce qui peut tre plus intuitif pour certaines applications. Un MAPE plus faible indique de meilleures performances du modle.

```python
### Prophet : forcasting time series package: sesonality, Holidays, special Events

import pandas as pd
from prophet import Prophet
from prophet.diagnostics import cross_validation
from datetime import datetime
path = "https://raw.githubusercontent.com/alimanager/smartBuilding/master/"
humd=pd.read_csv(f'{path}06_rhu_in.csv')
humd = humd["o4 [%]"]
humd = humd.rename(columns = {'timestamp [dd/mm/yyyy HH:MM]' : 'ds','o4 [%]': 'y'})
humd["ds"]= pd.to_datetime(humd["ds"], format = "%d/%m/%Y %H:%M")

humd.set_index('ds', inplace=True)
humd_hourly = humd.resample('H').mean()  # or use .sum(), .max(), etc. depending on your needs
humd_hourly.reset_index(inplace=True)

#Train & test split data 
Date_split = '2013-10-01'
train = humd[humd['ds'] <= Date_split] 
test = humd[humd['ds'] > Date_split] 

# Define initial and horizon
# Specify the length of the initial training period and the prediction horizon
initial = str((train['ds'].max() - train['ds'].min()).days) + ' days'
horizon = str((test['ds'].max() - test['ds'].min()).days) + ' days'

m = Prophet(changepoint_prior_scale='0.01',seasonality_prior_scale = '0.01',
            seasonality_mode='multiplicative',
            yearly_seasonality=True, weekly_seasonality=True,
            changepoint_range=0.9)
# m = Prophet()
m.add_country_holidays(country_name='AT') ### country_name='FR'for france
m.fit(humd)
# m.fit(humd)

future = m.make_future_dataframe(periods=30*24, freq='H')

# Prdiction & Vizualisation : 
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
fig = m.plot(forecast)
fig = m.plot_components(forecast)

# Evaluation : 
from prophet.diagnostics import performance_metrics
cv = cross_validation(model=m,initial=initial, horizon= horizon,period='90 days')
metrics = performance_metrics(cv)
print(metrics)
```

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Bidirectional


input_dim = X_train.shape[1]  # Replace this with the actual number of input features
X_val, y_val = X_test_seq, y_test_seq

# Crer une liste de modles
models = [
    # Sequential([
    #     Dense(64, input_dim=input_dim, activation='relu'),
    #     Dense(32, activation='relu'),
    #     Dense(1, activation='sigmoid')
    # ]),
    Sequential([
        Conv1D(32, kernel_size=3, activation='relu', input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
        MaxPooling1D(pool_size=2),
        Flatten(),
        Dense(1, activation='sigmoid')
    ]),
    Sequential([
        LSTM(64, input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
        Dense(1, activation='sigmoid')
    ]),
    Sequential([
        Bidirectional(LSTM(64), input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
        Dense(1, activation='sigmoid')
    ])
]

# Crer des listes pour stocker les mtriques
mse_scores = []
rmse_scores = []
r_squared_scores = []

# Boucle  travers les modles
for model in models:
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mse'])
    history = model.fit(X_train_seq, y_train_seq, epochs=10, batch_size=32, validation_data=(X_test_seq, y_test_seq))
    
    # valuer le modle sur l'ensemble de test
    loss, mse = model.evaluate(X_test_seq, y_test_seq)
    
    # Calculer RMSE et R-squared
    y_pred = model.predict(X_test_seq)
    rmse = np.sqrt(mean_squared_error(X_test_seq, y_pred))
    r_squared = 1 - (mean_squared_error(X_test_seq, y_pred) / np.var(X_test_seq))
    
    # Enregistrer les mtriques dans les listes
    mse_scores.append(mse)
    rmse_scores.append(rmse)
    r_squared_scores.append(r_squared)

# Trouver le meilleur modle en fonction de la mtrique souhaite (par exemple, MSE)
best_model_index = np.argmin(mse_scores)
best_model = models[best_model_index]

# Afficher les mtriques pour chaque modle
for i, model in enumerate(models):
    print(f"Model {i+1}:")
    print(f"MSE: {mse_scores[i]}")
    print(f"RMSE: {rmse_scores[i]}")
    print(f"R-squared: {r_squared_scores[i]}")
    print()

# Afficher les mtriques du meilleur modle
print("Best Model (based on MSE):")
print(f"MSE: {mse_scores[best_model_index]}")
print(f"RMSE: {rmse_scores[best_model_index]}")
print(f"R-squared: {r_squared_scores[best_model_index]}")

# Entraner le meilleur modle sur l'ensemble complet de donnes
best_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mse'])
best_history = best_model.fit(X_train_seq, y_train_seq, epochs=10, batch_size=32, validation_data=(X_test_seq, y_test_seq))

# Afficher les courbes d'apprentissage pour le meilleur modle
plt.plot(best_history.history['mse'], label='Training MSE')
plt.plot(best_history.history['val_mse'], label='Validation MSE')
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.legend()
plt.show()
```
