##### Normalisation des variables explicatives :
La normalisation est une technique souvent appliquée dans le cadre de la préparation des données pour l'apprentissage automatique. L'objectif de la normalisation est de re-échelonner les valeurs des colonnes numériques dans l'ensemble de données sans déformer les différences dans les gammes de valeurs.

La normalisation est souvent utilisée pour garantir que toutes les données d'un ensemble de données ont la même échelle, empêchant une seule variable de dominer le modèle. Par exemple, si une variable se situe dans la plage de 0 à 1 et une autre dans la plage de 1 à 1000, le modèle pourrait être excessivement influencé par la plage plus grande. La normalisation résout ce problème en transformant toutes les variables à une échelle commune.

Il existe plusieurs méthodes courantes de normalisation, notamment :

* Normalisation Min-Max : Cette méthode rééchelonne les données à une plage fixe, généralement de 0 à 1.
* Échelonnage décimal : Cette méthode divise par la valeur absolue maximale de chaque valeur puis multiplie par 10.
* Normalisation du score Z (standardisation) : Cette méthode standardise les valeurs pour avoir une moyenne de 0 et un écart type de 1.

- pour les variables descripteurs, comme l'humidité, la temperature, la radiation globale. on utilise la standarisation pour raméne à moyenne 0 et une écart-type 1. La normalisation est une partie importante de la prétraitement des données en apprentissage automatique et peut avoir un impact significatif sur la performance d'un modèle. Cependant, elle n'est pas toujours nécessaire et peut parfois même être indésirable, selon le contexte.

##### Colinearty des variablies :

La multicolinéarité est un phénomène dans lequel une variable caractéristique dans l'ensemble de données est fortement corrélée linéairement avec une autre variable caractéristique dans le même ensemble de données. Cela peut être problématique dans un modèle de régression car cela peut exagérer l'importance des variables et rendre difficile la détermination de l'effet de chaque variable indépendamment. Il existe plusieurs façons de traiter la multicolinéarité.

Une matrice de corrélation est une table qui montre le coefficient de corrélation entre différentes variables ( 1 : une corrélation positive forte, 1 : une corrélation négative forte, 0 : l'absence de corrélation). Chaque cellule de la table montre la corrélation entre deux variables spécifiques. Une matrice de corrélation est utilisée pour résumer les données, comme une entrée dans des modèles plus avancés, et comme un outil de diagnostic pour les corrélations linéaires.

##### Optimisation de la Précision pour Minimiser les Faux Positifs dans la Prédiction de l'Occupation des Pièces: 

Lorsque nous évoquons la notion de pénalisation des faux positifs, nous mettons l'accent sur la nécessité de minimiser les instances où notre modèle prédit de manière incorrecte la présence d'occupants. Dans notre contexte, la classe positive correspond à la "présence d'occupation", signifiant que le modèle prédit l'existence d'une présence humaine dans la pièce.

Un faux positif se produit donc lorsque le modèle anticipe une présence humaine dans la pièce (présence d'occupation), alors qu'en réalité, l'espace est vacant. Ce scénario est indésirable car il peut entraîner une consommation d'énergie superflue (par exemple, si l'éclairage ou le chauffage restent activés du fait que le système interprète la pièce comme étant occupée), ou il peut potentiellement créer des problèmes de sécurité.

Par conséquent, l'objectif est de configurer notre modèle de manière à minimiser ces prédictions de faux positifs. La précision (Precision) est un indicateur pertinent à cet égard. Elle représente le rapport des observations positives correctement prédites au total des prédictions positives. Une précision élevée indique un faible taux de faux positifs.

L'optimisation de la précision signifie que le modèle sera adapté pour améliorer sa capacité à prédire correctement la classe positive et réduire le nombre de faux positifs. Cependant, cette approche ne prend pas en compte les faux négatifs, qui, dans notre cas, se produiraient lorsque le modèle prédit que la pièce est vide, mais qu'en réalité une présence humaine est bien là. Pour prendre en compte cette situation, nous pouvons utiliser un autre indicateur, le rappel (Recall), qui mesure la capacité du modèle à trouver tous les exemples positifs.

Selon les exigences spécifiques de notre système, nous pourrions envisager d'optimiser d'autres indicateurs ou une combinaison d'indicateurs, afin d'équilibrer le compromis entre les faux positifs et les faux négatifs.

##### la définition des métriques d'Evaluation : 
- Accuracy : est la proximité des résultats de mesure avec la valeur réelle.[Wiki](https://fr.wikipedia.org/wiki/Exactitude_et_pr%C3%A9cision)

- AUC : L'aire sous le courbe ROC: caractéristique de performance (d'un test) ou courbe sensibilité/spécificité, est une mesure de la performance d'un classificateur binaire.[ROC Wiki](https://en.wikipedia.org/wiki/Receiver_operating_characteristic?oldid=375634814)

- Prec. (Precision) :  La précision montre dans quelle mesure la répétabilité (ou reproductibilité) dans des conditions inchangées affine les résultats.[Wiki](https://fr.wikipedia.org/wiki/Exactitude_et_pr%C3%A9cision)

- F1 : combine les mesures de précision et rappel, elles-mêmes basées sur les taux de vrais positifs, faux positifs et faux négatifs. L'idée de la F-mesure est de s'assurer qu'un classificateur fait de bonnes prédictions de la classe pertinente (bonne précision) en suffisamment grand nombre (bon rappel) sur un jeu de données cible. [F1 wiki](https://fr.wikipedia.org/wiki/F-mesure)

- Kappa : est la proportion de l'accord entre codeurs et Pr(e) la probabilité d'un accord aléatoire. Si les codeurs sont totalement en accord, κ = 1. S'ils sont totalement en désaccord (ou en accord dû uniquement au hasard)[Kappa Wiki](https://fr.wikipedia.org/wiki/Kappa_de_Cohen)

- MCC :Le coefficient de corrélation de Matthews (MCC, pour Matthews correlation coefficient en anglais) est une mesure de la qualité des classifications binaires. Il prend des valeurs entre -1 et +1. Un coefficient de +1 représente une prédiction parfaite, 0 une prédiction moyenne aléatoire, et -1 une prédiction inverse.Il est généralement considéré comme une mesure équilibrée car il tient compte des vrais et faux positifs et négatifs, et n'est pas trompé par les classes déséquilibrées.[Phi Coefficient](https://en.wikipedia.org/wiki/Phi_coefficient)

Définition des arbres de décision : 

C'est une technique d'apprentissage supervisé : on utilise un ensemble de données pour lesquelles on connaît la valeur de la variable-cible afin de construire l'arbre (données dites étiquetées), puis on extrapole les résultats à l'ensemble des données de test. Les arbres de décision font partie des algorithmes les plus populaires en apprentissage automatique.[Arbre de décision wiki](https://fr.wikipedia.org/wiki/Arbre_de_d%C3%A9cision_(apprentissage)#cite_note-11)
